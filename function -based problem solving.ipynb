#Smart Parking Lot Management System


```python
def smart_parking(capacity, logs):
    current = 0
    peak = 0

    for log in logs:
        if log == "IN":
            current += 1
        elif log == "OUT":
            current -= 1
        # Track peak usage
        if current > peak:
            peak = current
    print("Currently Parked Vehicles:", current)
    if current > capacity:
        print("Parking Status: Exceeded Capacity")
    else:
        print("Parking Status: Available ")
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
smart_parking(capacity, logs)
```

#Online Food Delivery Time Estimator

```python
def estimate_delivery_time(distance, traffic, weather):
    time = distance * 5   # base time

    if traffic == "High":
        time += 10
    elif traffic == "Medium":
        time += 10

    if weather == "Rainy":
        time += 5
    elif weather == "Storm":
        time += 10

    return time
#input
distance = 8
traffic = "High"
weather = "Rainy"

eta = estimate_delivery_time(distance, traffic, weather)
print("Estimated Delivery Time:", eta, "minutes")
```

#Movie Theatre Seat Occupancy 

```python
def analyze_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"
    return occupancy, status
#  Input
total_seats = 200
booked_seats = [1,] *150
occupancy, status = analyze_occupancy(total_seats, booked_seats)
print("Occupancy:", int(occupancy), "%")
print("Show Status:", status)
```

#Cloud Server Load Classification System

```python
def server_load(cpu_readings):
    total=0
    for value in cpu_readings:
        total += value
    average = total / len(cpu_readings)
   
#  Input
cpu = [45, 60, 70, 85, 90]
avg, status = server_load(cpu)
print("Average CPU Load:", int(avg), "%")
print("Server Status:", status)
```

#Smart Classroom Resource Usage Monitor

```python
def monitor_resources(resource_usage):
    overused = []
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)
    alert = "Yes" if overused else "No"
    return overused, alert
#  Input
usage = { "Projector": 6, "AC": 9,"Lights": 4}
overused, alert = monitor_resources(usage)
print("Overused Resources:", ", ".join(overused))
print("Energy Alert:", alert)
```

#Online Event Registration Capacity Controller

```python
def manage_registrations(capacity, registrations):
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
    return confirmed, waitlisted, status
# Input
capacity = 100
registrations = 105
confirmed, waitlisted, status = manage_registrations(capacity, registrations)
print("Confirmed Registrations:", confirmed)
print("Waitlisted Users:", waitlisted)
print("Registration Status:", status)
```
