#Smart Light Controller

```python
class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"
    def turn_on(self):
        self.status = "ON"
        print(f"{self.light_name} is {self.status}")
    def turn_off(self):
        self.status = "OFF"
        print(f"{self.light_name} is {self.status}")
#  Input
light = SmartLight("Bedroom Light")
light.turn_on()
```

# Employee ID Card System


```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    def display_id_card(self):
        print("Employee ID Card")
        print("Name:",self.name)
        print("ID:",self.emp_id)
        print("Department:",self.department)
#  Input
emp = Employee("Rahul","EMP102","IT")
emp.display_id_card()
```

# Mobile Contact Record


```python
class contacts:
    def __init__(self,Contact_Name,Phone_Number):
        self.Contact_Name= Contact_Name
        self.Phone_Number=Phone_Number
    def display_contact(self):
        print("Contact Saved")
        print("Name:",self.Contact_Name)
        print("Phone:",self.Phone_Number)
#input
detail = contacts("Anita","9876543210")
detail.display_contact()
    
```

#Product Price Tag Generator

```python
class products:
    def __init__(self,Product_name,Price):
        self.Product_name=Product_name
        self.Price=Price
    def prints_tag(self):
        print("Product:",self.Product_name)
        print("Price: $",self.Price)
#input
emp=products("Headphones","2499")
emp.prints_tag()
```

#Movie Rating Display System

```python
class theatre:
    def __init__(self,Movie_Name,Rating):
        self.Movie_Name= Movie_Name
        self.Rating=Rating
    def display_movie_details(self):
        print("Name:",self.Movie_Name)
        print("Rating:",self.Rating,"/5")
#input
detail= theatre("Inception","4.8")
detail.display_movie_details()
```
